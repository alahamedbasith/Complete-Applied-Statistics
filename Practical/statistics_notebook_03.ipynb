## Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# loc=50 sets the mean of the distribution.
# scale=10 sets the standard deviation of the distribution.
# size=50 generates 50 random numbers.

# Generate 50 random numbers from a normal distribution with mean=50 and std=10
data = np.random.normal(loc=50, scale=10, size=50)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

We get normal distribution by applying numpy random normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate 50 random numbers
data = np.random.rand(50)

# Original data
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Log Transformation
data_log = np.log(data + 1)
plt.hist(data_log, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Log Transformed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Square Root Transformation
data_sqrt = np.sqrt(data + 1)
plt.hist(data_sqrt, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Square Root Transformed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Box-Cox Transformation
data_boxcox, lambda_ = stats.boxcox(data + 1)
plt.hist(data_boxcox, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Box-Cox Transformed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Yeo-Johnson Transformation
data_yeojohnson, lambda_ = stats.yeojohnson(data)
plt.hist(data_yeojohnson, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Yeo-Johnson Transformed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# Plot histograms of original and transformed data
def plot_histograms(data, transformed_data=None, title=''):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
    plt.title('Original Data')
    plt.xlabel('Value')
    plt.ylabel('Density')

    if transformed_data is not None:
        plt.subplot(1, 2, 2)
        plt.hist(transformed_data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
        plt.title(title)
        plt.xlabel('Value')
        plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Plot original data
plot_histograms(data)

# Apply Box-Cox transformation
data_boxcox, lambda_ = stats.boxcox(data + 1)
plot_histograms(data, data_boxcox, 'Box-Cox Transformed Data')

# Apply Yeo-Johnson transformation
data_yeojohnson, lambda_ = stats.yeojohnson(data)
plot_histograms(data, data_yeojohnson, 'Yeo-Johnson Transformed Data')

# Q-Q plot for normality test
def qq_plot(data, title=''):
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.show()

# Q-Q plot for original data
qq_plot(data, 'Q-Q Plot of Original Data')

# Q-Q plot for Box-Cox transformed data
qq_plot(data_boxcox, 'Q-Q Plot of Box-Cox Transformed Data')

# Q-Q plot for Yeo-Johnson transformed data
qq_plot(data_yeojohnson, 'Q-Q Plot of Yeo-Johnson Transformed Data')

# Shapiro-Wilk Test for normality
def normality_test(data):
    stat, p = stats.shapiro(data)
    print(f'Shapiro-Wilk Test: Statistic={stat:.3f}, p-value={p:.3f}')
    
print('Normality Test for Original Data:')
normality_test(data)

print('Normality Test for Box-Cox Transformed Data:')
normality_test(data_boxcox)

print('Normality Test for Yeo-Johnson Transformed Data:')
normality_test(data_yeojohnson)
```

## Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate uniformly distributed data
data = np.random.uniform(low=0, high=10, size=1000)

# Plot histogram of data
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram of Data')

# Plot the PDF of the uniform distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.ones_like(x) / (xmax - xmin)
plt.plot(x, p, 'k', linewidth=2, label='Uniform PDF')

plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Transforming data from a uniform distribution to a normal distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate sample data from a uniform distribution
uniform_data = np.random.uniform(0, 1, 1000)

# Apply the inverse CDF (Quantile Function) of the normal distribution
normal_data = stats.norm.ppf(uniform_data)

# Plot histograms to compare distributions
plt.figure(figsize=(12, 6))

# Histogram of uniform data
plt.subplot(1, 2, 1)
plt.hist(uniform_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Histogram of transformed normal data
plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='b')
plt.title('Transformed Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

## Standard Normal Distribution
The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1.

```python
# Standard normal distribution
data_standard = np.random.standard_normal(1000)

# Plot
plt.hist(data_standard, bins=30, density=True, alpha=0.6, color='b')
plt.title('Standard Normal Distribution')
plt.show()
```

## Z-Score
A Z-score indicates how many standard deviations an element is from the mean.

```python
# Example data
data = [2, 4, 6, 8, 10]
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-scores
z_scores = [(x - mean) / std_dev for x in data]
print("Z-scores:", z_scores)
```

## log-normal distribution
A log-normal distribution is a probability distribution of a random variable whose logarithm is normally distributed.

```python
# Parameters
mu = 0
sigma = 0.1

# Generate data
data_log_normal = np.random.lognormal(mu, sigma, 1000)

# Plot
plt.hist(data_log_normal, bins=30, density=True, alpha=0.6, color='r')
plt.title('Log-Normal Distribution')
plt.show()
```

The log-normal distribution is useful in simple terms because it helps model and understand data that has the following characteristics:

### **1. Skewed Data**

- **Description:** When data is not evenly spread but has a long tail on one side (usually the right).
- **Example:** Income distribution where most people earn a modest income, but a few people earn a lot more.

### **2. Non-Negative Values**

- **Description:** When the data can only be positive.
- **Example:** The price of a product, where prices can’t be negative.

### **3. Multiplicative Growth**

- **Description:** When the growth of the data is due to the product of multiple factors.
- **Example:** The growth of an investment where returns are compounded over time.

### **Example Use**

Imagine you want to model the growth of investments or the distribution of product prices. The log-normal distribution is helpful because it can represent data where:
- Most values are small or moderate.
- A few values are much larger.


### **In Summary**

The log-normal distribution is handy for understanding and predicting data that:
- Is positively skewed.
- Can’t be negative.
- Grows multiplicatively.

This makes it particularly useful for modeling financial data, product prices, and other similar types of data.

## Bernoulli Distribution
The Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability p and the value 0 with probability 1-p.

```python
from scipy.stats import bernoulli

# Parameter
p = 0.6

# Generate data
data_bernoulli = bernoulli.rvs(p, size=1000)

# Plot
plt.hist(data_bernoulli, bins=3, density=True, alpha=0.6, color='y')
plt.title('Bernoulli Distribution')
plt.show()
```

## Binomial Distribution

The binomial distribution is a discrete probability distribution of the number of successes in a sequence of n independent experiments.

```python
from scipy.stats import binom

# Parameters
n = 10
p = 0.5

# Generate data
data_binomial = binom.rvs(n, p, size=1000)

# Plot
plt.hist(data_binomial, bins=n+1, density=True, alpha=0.6, color='m')
plt.title('Binomial Distribution')
plt.show()
```
