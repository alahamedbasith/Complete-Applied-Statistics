## Central Limit Theorem (CLT)
The CLT states that the distribution of the sum (or average) of a large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution, regardless of the original distribution of the variables.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a uniform distribution
samples = np.random.uniform(low=0, high=1, size=(10000, 50))
samples
````


This line generates a NumPy array of random samples drawn from a uniform distribution. Here's a detailed explanation of each part:

1. **`np.random.uniform(low=0, high=1, size=(10000, 50))`**:
   - **`np.random.uniform`**: This is a function from the NumPy library that generates random numbers from a uniform distribution.
   - **`low=0`**: This specifies the lower bound of the distribution. The random numbers generated will be greater than or equal to 0.
   - **`high=1`**: This specifies the upper bound of the distribution. The random numbers generated will be less than 1.
   - **`size=(10000, 50)`**: This specifies the shape of the output array. Here, it creates an array with 10,000 rows and 50 columns, resulting in 500,000 random numbers in total.

In summary, the line of code creates a 2D NumPy array named `samples` with dimensions 10,000 by 50, where each element is a random number uniformly distributed between 0 (inclusive) and 1 (exclusive). This array can be used to study the Central Limit Theorem by computing the means of these samples.

````python
# Compute the mean of each sample
sample_means = np.mean(samples, axis=1)

# Plot the distribution of sample means
plt.hist(sample_means, bins=50, density=True)
plt.title("Central Limit Theorem")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
````

## Chebyshev Inequality
Chebyshev's Inequality provides a way to determine the minimum proportion of values that lie within a certain number of standard deviations from the mean for any random variable, regardless of its distribution.

Here is the statement of Chebyshev's Inequality:

\[ P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2} \]

where:
- \( X \) is a random variable with mean \( \mu \) and standard deviation \( \sigma \).
- \( k \) is a positive number.
- \( P(|X - \mu| \geq k\sigma) \) is the probability that the value of \( X \) is at least \( k \) standard deviations away from the mean.

To find the number of standard deviations \( k \) for a given probability, you can rearrange the inequality:

\[ \frac{1}{k^2} \leq \text{desired probability} \]

In this example:
- We generate a large number of samples from a uniform distribution between 0 and 10.
- We calculate the mean and standard deviation of the samples.
- We use Chebyshev's Inequality to determine the probability that a sample is at least \( k \) standard deviations away from the mean.
- We plot the distribution and indicate the mean and \( k \) standard deviations.

This demonstrates how Chebyshev's Inequality can be applied to any distribution, not just the Gaussian distribution.

````python
import numpy as np

# Define the parameters
desired_probability = 0.05  # Desired probability (e.g., 5%)

# Calculate the number of standard deviations k
k = np.sqrt(1 / desired_probability)

print(f"For a probability of {desired_probability}, the number of standard deviations k is approximately {k:.2f}")
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a uniform distribution
samples = np.random.uniform(low=0, high=10, size=10000)

# Calculate mean and standard deviation
mean = np.mean(samples)
std_dev = np.std(samples)

# Define the number of standard deviations k
k = 2  # Example: 2 standard deviations

# Calculate Chebyshev's inequality probability
chebyshev_prob = np.mean(np.abs(samples - mean) >= k * std_dev)
upper_bound = 1 / k**2

print(f"Chebyshev's probability for k={k}: {chebyshev_prob:.4f}")
print(f"Upper bound according to Chebyshev's Inequality: {upper_bound:.4f}")

# Plot the distribution with mean and k standard deviations
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean - k*std_dev, color='b', linestyle='dashed', linewidth=2, label=f'-{k} Std Dev')
plt.axvline(mean + k*std_dev, color='b', linestyle='dashed', linewidth=2, label=f'+{k} Std Dev')
plt.legend()
plt.title("Uniform Distribution with Chebyshev's Inequality")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

## Pareto Distribution
The Pareto distribution is a power-law probability distribution used to describe phenomena in various fields such as economics, sociology, and geophysics.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate Pareto distributed samples
shape, scale = 3, 1  # shape (alpha) and scale parameters
samples = (np.random.pareto(shape, 1000) + 1) * scale

# Plot the distribution
plt.hist(samples, bins=50, density=True)
plt.title("Pareto Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

## Exponential Distribution

The exponential distribution describes the time between events in a Poisson process, i.e., a process in which events occur continuously and independently at a constant average rate.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate exponential distributed samples
scale = 1  # scale parameter (lambda)
samples = np.random.exponential(scale=scale, size=1000)

# Plot the distribution
plt.hist(samples, bins=50, density=True)
plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

In the context of probability distributions, the term "scale" generally refers to a parameter that stretches or compresses the distribution along the horizontal axis (the values that the random variable can take). Here is how the scale parameter functions in the Pareto and Exponential distributions:

### Pareto Distribution

The Pareto distribution is often used to describe phenomena with heavy tails, such as income distribution or file sizes on the internet. 

- **Shape parameter (α or `shape`):** This parameter controls the "heaviness" of the tail. Higher values of α result in a less heavy tail.
- **Scale parameter (x_m or `scale`):** This is the minimum value that the random variable can take. It shifts the distribution along the x-axis.

In mathematical terms, the probability density function (PDF) of a Pareto distribution is:

\[ f(x) = \frac{\alpha x_m^\alpha}{x^{\alpha + 1}} \text{ for } x \geq x_m \]

where \( \alpha > 0 \) and \( x_m > 0 \).

### Exponential Distribution

The Exponential distribution is often used to model the time between events in a Poisson process (i.e., events occurring continuously and independently at a constant average rate).

- **Rate parameter (λ or `rate`):** This is the reciprocal of the scale parameter. It represents the rate at which events occur. Higher values of λ mean events occur more frequently.
- **Scale parameter (β or `scale`):** This is the inverse of the rate parameter, \( \beta = 1 / \lambda \). It represents the mean time between events. Higher values of β mean that events occur less frequently.

The probability density function (PDF) of an Exponential distribution is:

\[ f(x) = \lambda e^{-\lambda x} \text{ for } x \geq 0 \]

or equivalently,

\[ f(x) = \frac{1}{\beta} e^{-x/\beta} \text{ for } x \geq 0 \]

where \( \lambda > 0 \) and \( \beta > 0 \).

## Poisson Distribution

The Poisson distribution describes the number of events occurring within a fixed interval of time or space, given a constant mean rate of occurrence.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate Poisson distributed samples
lambda_ = 5  # rate parameter
samples = np.random.poisson(lam=lambda_, size=1000)

# Plot the distribution
plt.hist(samples, bins=30, density=True)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
````

Sure, let's delve into a more detailed real-time example using a customer service center scenario:

### Scenario

1. **Poisson Distribution:**
   - We will model the number of calls received by the customer service center in one hour.
   - Assume the center receives an average of 10 calls per hour.

2. **Exponential Distribution (Interarrival Times):**
   - We will model the time between consecutive calls.
   - With an average of 10 calls per hour, the average time between calls is 6 minutes (0.1 hours).

### Poisson Distribution: Number of Calls per Hour

In this part, we'll use the Poisson distribution to simulate the number of calls received in one hour.


````python
#### Python Example
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_ = 10  # Average number of calls per hour

# Generate Poisson distributed samples (number of calls per hour)
num_calls_per_hour = np.random.poisson(lam=lambda_, size=1000)

# Plot the distribution
plt.hist(num_calls_per_hour, bins=30, density=True, alpha=0.75, color='blue')
plt.title("Poisson Distribution: Number of Calls per Hour")
plt.xlabel("Number of Calls")
plt.ylabel("Frequency")
plt.show()

# Display mean and standard deviation
mean_calls = np.mean(num_calls_per_hour)
std_dev_calls = np.std(num_calls_per_hour)
print(f"Mean number of calls per hour: {mean_calls:.2f}")
print(f"Standard deviation: {std_dev_calls:.2f}")
````

### Exponential Distribution: Time Between Consecutive Calls

In this part, we'll use the Exponential distribution to simulate the time between consecutive calls.


````python
#### Python Example
import numpy as np
import matplotlib.pyplot as plt

# Parameters
rate = 10  # Rate parameter (calls per hour)
scale = 1 / rate  # Scale parameter for the exponential distribution (hours per call)

# Generate exponential distributed interarrival times (time between consecutive calls)
interarrival_times = np.random.exponential(scale=scale, size=1000)

# Plot the distribution of interarrival times
plt.hist(interarrival_times, bins=30, density=True, alpha=0.75, color='green')
plt.title("Exponential Distribution: Time Between Consecutive Calls")
plt.xlabel("Interarrival Time (hours)")
plt.ylabel("Frequency")
plt.show()

# Cumulative sum of interarrival times to get event times
event_times = np.cumsum(interarrival_times)

# Plot the cumulative number of events over time
plt.plot(event_times, np.arange(1, len(event_times) + 1), marker='o', linestyle='none')
plt.title("Cumulative Number of Calls Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Number of Calls")
plt.show()

# Display mean and standard deviation of interarrival times
mean_interarrival_time = np.mean(interarrival_times)
std_dev_interarrival_time = np.std(interarrival_times)
print(f"Mean interarrival time: {mean_interarrival_time:.2f} hours")
print(f"Standard deviation: {std_dev_interarrival_time:.2f} hours")
````

### Explanation

1. **Poisson Distribution: Number of Calls per Hour:**
   - We generate 1,000 samples from a Poisson distribution with a rate parameter \( \lambda = 10 \) (average number of calls per hour).
   - We plot a histogram to visualize the distribution of the number of calls received per hour.
   - We calculate and display the mean and standard deviation of the number of calls.

2. **Exponential Distribution: Time Between Consecutive Calls:**
   - We generate 1,000 samples from an Exponential distribution with a scale parameter \( \beta = 0.1 \) hours (average time between calls).
   - We plot a histogram to visualize the distribution of interarrival times.
   - We calculate the cumulative sum of interarrival times to get the actual event times and plot these times to show the cumulative number of calls over time.
   - We calculate and display the mean and standard deviation of the interarrival times.

To find the probability that exactly 6 trains will arrive in a two-hour period given that trains arrive at an average rate of 4 trains per hour, we can use the Poisson distribution. 

### Poisson Distribution Formula
The Poisson distribution gives the probability of a given number of events occurring in a fixed interval of time or space. The probability of observing \( k \) events in an interval is given by:

\[ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \]

where:
- \( \lambda \) is the average number of events in the interval.
- \( k \) is the actual number of events we want to find the probability for.
- \( e \) is the base of the natural logarithm.

### Given:
- The average arrival rate \( \lambda \) is 4 trains per hour.
- For a two-hour period, the average arrival rate becomes \( 2 \times 4 = 8 \) trains.
- We want to find the probability of exactly 6 trains arriving in a two-hour period (\( k = 6 \)).

### Python Example
We'll use Python to calculate this probability using the Poisson distribution from the `scipy.stats` module.

```python
import scipy.stats as stats

# Parameters
lambda_2_hours = 8  # Average number of trains in 2 hours
k = 6  # Number of trains

# Calculate the probability of exactly 6 trains arriving in 2 hours
probability = stats.poisson.pmf(k, lambda_2_hours)

print(f"The probability of exactly {k} trains arriving in a two-hour period is: {probability:.4f}")
```

### Explanation
1. **Parameter Calculation:**
   - We calculate the average number of trains in a two-hour period (\( \lambda = 8 \)).

2. **Probability Calculation:**
   - We use the `poisson.pmf` function from the `scipy.stats` module to calculate the probability mass function (PMF) for exactly \( k = 6 \) trains.

The Python code will output the probability of exactly 6 trains arriving in a two-hour period.

### Manual Calculation
For completeness, here is the manual calculation:

\[ P(X = 6) = \frac{8^6 e^{-8}}{6!} \]

\[ P(X = 6) = \frac{262144 \times e^{-8}}{720} \]

\[ P(X = 6) \approx \frac{262144 \times 0.00033546}{720} \]

\[ P(X = 6) \approx \frac{87.8902}{720} \]

\[ P(X = 6) \approx 0.122 \]

Thus, the probability of exactly 6 trains arriving in a two-hour period is approximately 0.122.
